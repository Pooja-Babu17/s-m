```python
#LEGENT
seaborn.move_legend(obj, loc, **kwargs)
obj: The object with the plot. This argument can be either a Seaborn object (e.g., seaborn.FacetGrid or seaborn.PairGrid) or a Matplotlib object (matplotlib.axes.Axes or matplotlib.figure.Figure).

loc: Location argument, specified as a string or integer, similar to the loc parameter in matplotlib.axes.Axes.legend(). It determines where the legend will be placed in the plot. Common string values include 'upper right', 'upper left', 'lower right', 'lower left', 'right', 'center left', 'center right', 'lower center', 'upper center', and 'center'. Integer values can also be used to represent specific locations.


seaborn.despine(fig=None, ax=None, top=True, right=True, left=False, bottom=False, offset=None, trim=False)kwargs: Other keyword arguments that are passed to matplotlib.axes.Axes.legend() for additional customization of the legend.
```

```python
seaborn.despine(fig=None, ax=None, top=True, right=True, left=False, bottom=False, offset=None, trim=False)


The seaborn.despine function in Seaborn is used to remove or adjust the appearance of the spines (the borders) of a plot. It provides a way to customize the visual style of the plot by controlling which spines are visible and their positions.


fig: A Matplotlib Figure instance. If provided, this specifies the figure on which to operate. If not provided, the current Figure is used.

ax: A Matplotlib Axes instance. If provided, this specifies the axes on which to operate. If not provided, the current Axes is used.

top, right, left, bottom: Boolean values indicating whether to remove the spines on the top, right, left, and bottom sides of the plot, respectively. If True, the spine is removed; if False, the spine is retained.

offset: A distance (in points) by which to offset the spines from the data area. This can be useful to make room for labels or other elements.

trim: If True, the spines are trimmed to the axis limits after offsetting. This can be useful when spines are offset to avoid overlapping with data points.
```

```python
from matplotlib.animation import FuncAnimation

# Assuming df is your DataFrame
fig, ax = plt.subplots()
x_data = df["hsc_p"]
y_data = df["ssc_p"]
line, = ax.plot(x_data, y_data)

def update(frame):
    line.set_xdata(x_data[:frame])
    line.set_ydata(y_data[:frame])
    return line,

ani = FuncAnimation(fig, update, frames=len(x_data), interval=100, blit=True)
plt.xlabel("Higher Secondary Percentage (hsc_p)")
plt.ylabel("Secondary Percentage (ssc_p)")
plt.title("Animated Plot of Secondary and Higher Secondary Percentages")
plt.show()
```
